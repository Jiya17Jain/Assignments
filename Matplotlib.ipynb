Matplotlib i a library in python used to plot graph . Various plots like scatter plot , line plot , bar plot , box plot , Histogram can be plotted using matplotlib . 

Ans2.

Scatter plot , a graph in which the values of two variables are plotted along two axes.

```python

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

Ans4.

Bar plot is used to depict dat between categorical data and numerical data.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

Ans5.

Box plot ,a simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.boxplot(box2)
```

Subplots method provides a way to plot multiple plots on a single figure.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,2,1)
plt.plot(x,y)
```
